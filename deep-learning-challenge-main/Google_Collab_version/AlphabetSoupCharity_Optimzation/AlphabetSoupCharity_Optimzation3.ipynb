```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
application_df = pd.read_csv("/content/charity_data.csv")
application_df.head()
```

```python

# Drop the non-beneficial ID columns, 'CLASSIFICATION'.

application_df = application_df.drop(columns = ['CLASSIFICATION'])
```

```python

# Determining the number of unique values in each column.
for x in application_df.columns:
    print(x, len(application_df[x].unique()))
```

```python
# Look at NAMES value counts for binning
name_count = application_df['NAME'].value_counts()
name_count
```

```python
# list of names to be replaced
names_replace = list(name_count[name_count<10].index)

for app in names_replace:
    application_df['NAME'] = application_df['NAME'].replace(app,"Other")

application_df['NAME'].value_counts()
```

```python
# Look at ORGANIZATION value counts for binning
organization_binning = application_df['ORGANIZATION'].value_counts()
organization_binning 
```

```python
# Convert categorical data to numeric
application_df = pd.get_dummies(application_df,dtype=float)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df.drop('IS_SUCCESSFUL', axis=1).values
y = application_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_features = X_train_scaled.shape[1]
hidden_nodes1=5
hidden_nodes2=10
hidden_nodes3=15

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes1, input_dim=input_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes2, activation='relu'))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes3, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,validation_split=0.15, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
